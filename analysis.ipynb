```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import sheet korba-urban from data.xlsx

df_korba_urban = pd.read_excel('data.xlsx', sheet_name='korba-urban')
df_korba_mines = pd.read_excel('data.xlsx', sheet_name='korba-mines')
df_raigarh_urban = pd.read_excel('data.xlsx', sheet_name='raigarh-urban')
df_raigarh_mines = pd.read_excel('data.xlsx', sheet_name='raigarh-mines')
df_surguja_urban = pd.read_excel('data.xlsx', sheet_name='surguja-urban')
df_surguja_mines = pd.read_excel('data.xlsx', sheet_name='surguja-mines')
df_koriya_urban = pd.read_excel('data.xlsx', sheet_name='koriya-urban')
df_koriya_mines = pd.read_excel('data.xlsx', sheet_name='koriya-mines')
```

## Urban

### Area Increase

```python

# plot line graph for area in all 4 cities as faded lines and average of all 4 cities as bold line

plt.figure(figsize=(7,7))
plt.plot(df_korba_urban['year'], df_korba_urban['area'], color='brown', alpha=0.5, label='korba')
plt.plot(df_raigarh_urban['year'], df_raigarh_urban['area'], color='blue', alpha=0.5, label='raigarh')
plt.plot(df_surguja_urban['year'], df_surguja_urban['area'], color='green', alpha=0.5, label='surguja')
plt.plot(df_koriya_urban['year'], df_koriya_urban['area'], color='orange', alpha=0.5, label='koriya')

# plot average of all 4 cities as bold line
average = (df_korba_urban['area'] + df_raigarh_urban['area'] + df_surguja_urban['area'] + df_koriya_urban['area'])/4
# plot the average line in bold along with writing the first and last value of average on the graph
plt.plot(df_korba_urban['year'], average, color='red', alpha=1, linewidth=3,marker='o', label='average')

# show labels = true
plt.legend()


# put labels for line colors
plt.xlabel('year')
plt.ylabel('Area (in sq. km)')
plt.title('Area in 4 cities (URBAN)')
plt.show()


print("Average area increased from ", average[0], " to ", average[len(average)-1], " in 4 cities (URBAN)")
```

### Max Summer Temperature Increase

```python
# write code to do analysis of data especially to show the rise in temperature over the years in all 4 cities

# plot line graph for max_summer in all 4 cities as faded lines and average of all 4 cities as bold line

plt.figure(figsize=(7,7))
plt.plot(df_korba_urban['year'], df_korba_urban['max_summer'], color='brown', alpha=0.5, label='korba')
plt.plot(df_raigarh_urban['year'], df_raigarh_urban['max_summer'], color='blue', alpha=0.5, label='raigarh')
plt.plot(df_surguja_urban['year'], df_surguja_urban['max_summer'], color='green', alpha=0.5, label='surguja')
plt.plot(df_koriya_urban['year'], df_koriya_urban['max_summer'], color='orange', alpha=0.5, label='koriya')

# plot average of all 4 cities as bold line
average = (df_korba_urban['max_summer'] + df_raigarh_urban['max_summer'] + df_surguja_urban['max_summer'] + df_koriya_urban['max_summer'])/4
# plot the average line in bold along with writing the first and last value of average on the graph
plt.plot(df_korba_urban['year'], average, color='red', alpha=1, linewidth=3,marker='o', label='average')

# show labels = true
plt.legend()


# put labels for line colors
plt.xlabel('year')
plt.ylabel('Max Summer Temperature')
plt.title('Max Summer Temperature in 4 cities (URBAN)')
plt.show()


print("Average Temperature increased from ", average[0], " to ", average[len(average)-1], " in 4 cities (URBAN)")
```

1. As we can see from the plot above, in all 4 cities, there have been substantial increase in maximum summer temperature over the years as the area of urbanization has increased. The highest temperature recorded is 50.11 degree celcius in Raigarh, which also has the highest increase of temperature : 17.64 degree celcius. 

2. The average temperature (red line) across cities has increased by 32 % from 32.6 to 43.1 degree celcius.

### Mean Winter Temperature Increase & Summer Heatmaps

```python
# creating a heatmap

temperature_array = np.array([df_korba_urban['mean_winter'], df_raigarh_urban['mean_winter'], df_surguja_urban['mean_winter'], df_koriya_urban['mean_winter']])
years = np.array(df_korba_urban['year'])
cities = np.array(['Korba', 'Raigarh', 'Surguja', 'Koriya'])

plt.figure(figsize=(15,7))

# Create the heatmap
fig, ax = plt.subplots()
heatmap = ax.imshow(temperature_array, cmap='hot_r')

# Set the tick labels and positions
ax.set_xticks(np.arange(len(years)))
ax.set_yticks(np.arange(len(cities)))
ax.set_xticklabels(years)
ax.set_yticklabels(cities)

# Rotate the tick labels and set alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add colorbar
cbar = ax.figure.colorbar(heatmap, ax=ax)
cbar.ax.set_ylabel('Temperature Rise', rotation=-90, va="bottom")

# Set axis labels
ax.set_xlabel('Years')
ax.set_ylabel('Cities')
ax.set_title('Mean Winter Temperature Rise (URBAN)')

# Display the plot
plt.show()

average_temperature_across_years = (df_korba_urban['mean_winter'] + df_raigarh_urban['mean_winter'] + df_surguja_urban['mean_winter'] + df_koriya_urban['mean_winter'])/4
# make a bar graph for average temperature across years

plt.figure(figsize=(5,5))
# y axis strating from 15 to 30
plt.ylim(15, 25)
# only show the years in x axis
plt.xticks(df_korba_urban['year'])
plt.bar(df_korba_urban['year'], average_temperature_across_years, color='red', alpha=1, linewidth=3, label='average')
plt.xlabel('year')
plt.ylabel('Average Winter Temperature')
plt.title('Average Winter Temperature across years (URBAN)')
plt.show()


```

```python
temperature_array = np.array([df_korba_urban['max_summer'], df_raigarh_urban['max_summer'], df_surguja_urban['max_summer'], df_koriya_urban['max_summer']])


plt.figure(figsize=(15,7))

# Create the heatmap
fig, ax = plt.subplots()
heatmap = ax.imshow(temperature_array, cmap='hot_r')

# Set the tick labels and positions
ax.set_xticks(np.arange(len(years)))
ax.set_yticks(np.arange(len(cities)))
ax.set_xticklabels(years)
ax.set_yticklabels(cities)

# Rotate the tick labels and set alignment
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add colorbar
cbar = ax.figure.colorbar(heatmap, ax=ax)
cbar.ax.set_ylabel('Temperature Rise', rotation=-90, va="bottom")

# Set axis labels
ax.set_xlabel('Years')
ax.set_ylabel('Cities')
ax.set_title('Max Summer Temperature Rise (URBAN)')

# Display the plot
plt.show()
```

## Increase in Temperature Differences

### Urban

```python
import matplotlib.pyplot as plt

average_summer_max_temps = (df_korba_urban['max_summer'] + df_raigarh_urban['max_summer'] + df_surguja_urban['max_summer'] + df_koriya_urban['max_summer'])/4
average_winter_min_temps = (df_korba_urban['min_winter'] + df_raigarh_urban['min_winter'] + df_surguja_urban['min_winter'] + df_koriya_urban['min_winter'])/4


plt.figure(figsize=(5, 5))

# Calculate the range for each year
ranges = [(winter, summer) for summer, winter in zip(average_summer_max_temps, average_winter_min_temps)]

# Create the range bar plot
for i, year in enumerate(years):
    plt.plot([year, year], ranges[i], marker='_', color='blue', markersize=10, linewidth=2)

# Set labels and title
# y range must be from 0 to 30
plt.ylim(0, 50)
plt.xticks(years)
plt.xlabel('Years')
plt.ylabel('Temperature Difference (Summer - Winter)')
plt.title('Widening Gap Between Extreme Temperatures : URBAN')

# Show the plot
plt.grid(True)
plt.show()
```
